```python
import numpy as np
import matplotlib.pyplot as plt
#import my_toolbox as my
from netCDF4 import Dataset
import sys
import cartopy
#sys.path.insert(0,'/Users/nbruggemann/syn_dropbox/sync/Promotion/src/pyicon')
import pyicon as pyic
import matplotlib
```

```python
ccrs_proj = cartopy.crs.PlateCarree()
```

```python
x = np.linspace(-180,180,360.)
y = np.linspace(-90,90,180.)
X, Y = np.meshgrid(x,y)
Z = np.cos(X*np.pi/180.)*np.sin(Y*np.pi/180.)
```

```python
hca, hcb = pyic.arrange_axes(2,2, asp=0.5, plot_cb=[0,1,0,1],
                       sharex=True, sharey=True,
                       projection=ccrs_proj,
                       )
for nn, (ax,cax) in enumerate(zip(hca,hcb)):
  #ax.set_title('axes %d'%(nn))
  #ax.set_title('vert. int. temp. bias')
  #ax.set_title('IDEMIX %d'%(nn), loc='right', fontsize=10)
  ax.set_title('IDEMIX %d: vert. int. temp. bias' % (nn))
  if cax!=0:
    cax.set_title('$^o$C')
  pyic.plot_settings(ax=ax, projection=ccrs_proj, template='global')
```
