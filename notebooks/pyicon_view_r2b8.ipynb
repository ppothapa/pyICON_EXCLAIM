```python
from matplotlib.patches import Rectangle
def add_sonett(ax):
    rectprop = dict(facecolor='none', edgecolor='k', lw=2)
    ax.add_patch(Rectangle((-12,-37.5), 5, 4, **rectprop))
    ax.text(-9.5, -33.5, 'SONETT II', ha='center', va='bottom', fontsize=8)

    rectprop = dict(facecolor='none', edgecolor='k', lw=2)
    ax.add_patch(Rectangle((1,-35), 9, 5, **rectprop))
    ax.text(5.5, -30., 'SONETT I', ha='center', va='bottom', fontsize=8)

    ax.text(-2.5, -35, 'Walvis Ridge', va='center', ha='center', rotation=45, fontsize=8)
    return
```

## Plotting on a regular grid

Define a dataset with giving the path to the data and information about the triangular grid and the regular grid. The regular grid needs to be created in advance (use pyic.calc_ckdtree for that).

```python
# run = 'exp.ocean_ncep6h_r2b8_20096-FZA'
run = 'exp.ocean_omip_long_tke_r2b8_20134-WWG'
runname = ''
gname = 'r2b8_oce_r0004'
lev = 'L128'

path_data     = f'/work/mh0033/m211054/projects/icon/icon-oes-1.3.01/experiments/{run}/outdata/'
# path_data     = f'/mnt/lustre01/work/mh0033/m300602/proj_vmix/icon/{runname}/icon-oes/experiments/{run}/'
path_grid     = f'/mnt/lustre01/work/mh0033/m300602/icon/grids/{gname}/'
path_ckdtree  = f'{path_grid}ckdtree/'
fpath_ckdtree = f'{path_grid}ckdtree/rectgrids/{gname}_res1.00_180W-180E_90S-90N.npz'
fpath_ckdtree_zave = fpath_ckdtree
fpath_fx      = f'{path_grid}{gname}_{lev}_fx.nc'
```

```python
IcD = pyic.IconData(
               fname        = run+'_????????????????.nc',
               path_data    = path_data,
               path_grid    = path_grid,
               gname        = gname,
               lev          = lev,
               do_triangulation = False,
               omit_last_file   = True,
               calc_coeff       = False,
               verbose          = True,
              )
```

Start the ncview-like plotting:

```python
PyicV = pyic.view.hplot(IcD, logplot=False)
add_sonett(PyicV.ax)
```

```python
PyicV = pyic.view.vplot(IcD, log2vax=False, logplot=False)
```
