Load the necessary modules:

```python
import pyicon_jupyter as jup
import pyicon as pyic      
%matplotlib notebook
```

## Plotting on a regular grid

Define a dataset with giving the path to the data and information about the triangular grid and the regular grid. The regular grid needs to be created in advance (use pyic.calc_ckdtree for that).

```python
path_data = '/mnt/lustre01/work/mh0033/m211054/projects/icon/solver/icon-oes-solver-HB/experiments/ocean_benchmark/outdata/'
IcD = pyic.IconData(
    fpath_grid_triangular  = path_data+'ocean_benchmark_20161011T000000Z.nc',
    #region                 = 'hurricane',
    fname_rgrid            = 'r2b9ocean_r0.05_100w_30w_2n_40n.npz',
    #fname_rgrid            = 'r2b9ocean_r0.3_180w_180e_90s_90n.npz',
    path_data              = path_data,
    search_str             = 'ocean_benchmark_????????????????.nc',
       )
```

Start the ncview-like plotting:

```python
IcD, Ip = jup.hplot(IcD)
```

## Plotting on the original grid

Similar as before: Define a dataset with giving the path to the data and information about the triangular grid.

```python
IcD = pyic.IconData(
       fpath_grid_triangular  = path_data+'ocean_benchmark_20161011T000000Z.nc',
       region                 = 'hurricane',
       path_data              = path_data,
       search_str             = 'ocean_benchmark_*.nc',
       lon_reg                = [-80, -75],
       lat_reg                = [16, 18],
       use_tgrid              = True,
       )
```

Start the ncview-like plotting:

```python
IcD, Ip = jup.hplot(IcD)
```
